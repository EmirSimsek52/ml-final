```python
from google.colab import files
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, roc_curve, roc_auc_score, classification_report
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
import numpy as np
```

```python
uploaded = files.upload()
veri = pd.read_csv("veri_seti.csv") #Veri dosyasını kaydediyoruz
print(veri.head()) # Yazdırıyoruz
```

```python
scaler = StandardScaler()
X = veri.iloc[:, :-1]
y = veri.iloc[:, -1]

# Veriyi normalize etme
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Normalize edilmiş veri setini DataFrame'e dönüştürme
normalized_df = pd.DataFrame(X_normalized, columns=veri.columns[:-1])
print(normalized_df)
```

```python
def evaluate_model(model, X, y, model_name):
    def plot_metrics(metrics):
        # Metrik isimleri ve değerleri
        metric_names = list(metrics.keys())
        metric_values = list(metrics.values())

        # Sütun grafiği çizimi
        plt.figure(figsize=(10, 6))
        plt.bar(metric_names, metric_values, color='skyblue')
        plt.xlabel('Metrics')
        plt.ylabel('Values')
        plt.title('Model Evaluation Metrics')
        plt.xticks(rotation=45, ha='right')
        plt.show()

    # Veriyi eğitim ve test setlerine ayırma
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

    # Modeli eğitme
    model.fit(X_train, y_train)

    # Tahmin yapma
    y_pred = model.predict(X_test)
    y_pred_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

    # Metriği hesaplama
    cm = confusion_matrix(y_test, y_pred)
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    report_dict = classification_report(y_test, y_pred, output_dict=True)
    sensitivity = report_dict["1"]["recall"]
    specificity = report_dict["0"]["recall"]

    roc_auc = roc_auc_score(y_test, y_pred_prob) if y_pred_prob is not None else None
    fpr, tpr, _ = roc_curve(y_test, y_pred_prob) if y_pred_prob is not None else (None, None, None)

    # Confusion Matrix grafiği
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'{model_name} Confusion Matrix')
    plt.show()

    # ROC Eğrisi grafiği
    if fpr is not None and tpr is not None:
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
        plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'{model_name} ROC Curve')
        plt.legend(loc="lower right")
        plt.show()

    # Metrikleri görselleştir
    plot_metrics({
        "Accuracy": accuracy,
        "F1 Score": f1,
        "Sensitivity": sensitivity,
        "Specificity": specificity,
    })

    # Sonuçları döndürme
    return {
        "confusion_matrix": cm,
        "accuracy": accuracy,
        "f1_score": f1,
        "sensitivity": sensitivity,
        "specificity": specificity,
        "roc_auc": roc_auc,
        "roc_curve": (fpr, tpr),
        "classification_report": report
    }
```

```python
from sklearn.naive_bayes import GaussianNB
import seaborn as sns


# Naive Bayes modelini değerlendirme
nb_model = GaussianNB()
nb_metrics = evaluate_model(nb_model, X_normalized, y, "Naive Bayes")

# Sonuçları yazdırma
print(f"Naive Bayes Doğruluk Oranı: {nb_metrics['accuracy']}")
print("Naive Bayes Sınıflandırma Raporu:")
print(nb_metrics['classification_report'])
print("Naive Bayes Karışıklık Matrisi:")
print(nb_metrics['confusion_matrix'])
```

```python
from sklearn.neighbors import KNeighborsClassifier

# K-NN modelinin optimizasyonu
k_range = range(1, 50)  # 1'den
param_grid_knn = {'n_neighbors': k_range}
knn = KNeighborsClassifier()
knn_grid = GridSearchCV(knn, param_grid_knn, cv=5, scoring='accuracy')
knn_grid.fit(X_normalized, y)
best_k = knn_grid.best_params_['n_neighbors']
knn_model = KNeighborsClassifier(n_neighbors=best_k)

# K-NN modelini değerlendirme
knn_metrics = evaluate_model(knn_model, X_normalized, y, f"K-NN (Best k={best_k})")
print(f"K-NN Doğruluk Oranı (En iyi k={best_k}): {knn_metrics['accuracy']}")
print("K-NN Sınıflandırma Raporu:")
print(knn_metrics['classification_report'])
print("K-NN Karışıklık Matrisi:")
print(knn_metrics['confusion_matrix'])

```

```python
# MLP sınıflandırıcısını uygulama ve sonuçları yazdırma
mlp_model = MLPClassifier(random_state=42, max_iter=300)
mlp_metrics = evaluate_model(mlp_model, X_normalized, y, "MLP")
print(f"MLP Doğruluk Oranı: {mlp_metrics['accuracy']}")
print("MLP Sınıflandırma Raporu:")
print(mlp_metrics['classification_report'])
print("MLP Karışıklık Matrisi:")
print(mlp_metrics['confusion_matrix'])
```

```python
# SVM sınıflandırıcısını uygulama ve sonuçları yazdırma
svm_model = SVC(probability=True, random_state=42)
svm_metrics = evaluate_model(svm_model, X_normalized, y, "SVM")
print(f"SVM Doğruluk Oranı: {svm_metrics['accuracy']}")
print("SVM Sınıflandırma Raporu:")
print(svm_metrics['classification_report'])
print("SVM Karışıklık Matrisi:")
print(svm_metrics['confusion_matrix'])
```
